```python
from sklearn.feature_selection import SelectPercentile
X = raw_data.drop(['rating', 'title', 'calories', 'protein', 'fat', 'sodium'], 1)
Y = raw_data.rating
select = SelectPercentile(percentile=4.4)
select.fit(X, Y)
X_select = select.transform(X)
print('x shape is: {}'.format(X_select.shape))
```

```python
mask = select.get_support()
print(mask)
plt.matshow(mask.reshape(1,-1), cmap='gray_r')
```

```python
X = raw_data.drop(['rating', 'title', 'calories', 'protein', 'fat', 'sodium'], 1)
X_new = select.transform(X)
```

```python
X_new.shape
```

```python
from sklearn.svm import SVC
svc = SVC()
Y = pd.Series(np.where(raw_data.rating > 3, 1, 0))
svc.fit(X_new,Y)
```

```python
plt.scatter(Y, svc.predict(X_new))
```

```python
svc.score(X_new, Y)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(svc, X_new, Y, cv=5)
```
